```python
import pandas as pd
```

```python
import numpy as np
```

```python
df = pd.read_csv('financial_loan.csv')
```

```python
df.head(10)
```

```python
df.info()
```

```python
dt_null = df.isnull().sum() / df.shape[0] * 100
dt_null
```

```python
df['loan_status'].groupby(df['loan_status']).count()
```

```python
date_cols = ['issue_date','last_credit_pull_date'	,'last_payment_date','next_payment_date']
```

```python
df[date_cols].head()
df[date_cols]
```

```python
for col in date_cols:
  df[col] = pd.to_datetime(df[col], dayfirst=True, format='mixed')
```

```python
df[date_cols].head()
```

```python
df['term_months'] = df['term'].str.extract(r"(\d+)").astype(int)
```

```python
df['term_months'].head()
```

```python
df[['term','term_months']].drop_duplicates()
```

```python
df["emp_length"].value_counts(dropna=False)
```

```python
def clean_emp_length(x):
    if pd.isna(x):
        return None
    x = x.strip().lower()
    if x == "n/a":
        return None
    if "<" in x:
        return 0.5
    if "10+" in x:
        return 10
    return int(x.split()[0])

df["emp_length_years"] = df["emp_length"].apply(clean_emp_length)
```

```python
df[["emp_length", "emp_length_years"]].drop_duplicates().sort_values("emp_length_years")
```

```python
df["interest_rate_pct"] = df["int_rate"] * 100
```

```python
df["is_default"] = df["loan_status"].apply(
    lambda x: 1 if x == "Charged Off" else 0
)
```

```python
df[["loan_status", "is_default"]].value_counts()
```

```python
df["end_date"] = df["last_payment_date"].fillna(df["last_credit_pull_date"])
df["end_date"].head()
```

```python
df["loan_age_months"] = (
    (df["end_date"] - df["issue_date"]).dt.days / 30
).round(1)
```

```python
df[["issue_date", "end_date", "loan_age_months"]].head()
```

```python
df[df["loan_age_months"] < 0].shape
```

```python
df.loc[df["loan_age_months"] < 0,
       ["issue_date", "last_payment_date", "last_credit_pull_date", "loan_status"]].head(10)
```

```python
df.loc[df["loan_age_months"] < 0, "loan_age_months"] = None
df[["issue_date", "end_date", "loan_age_months"]].head()
```

```python
df["loan_age_months"].describe()
```

```python
df["issue_year"] = df["issue_date"].dt.year
df["issue_month"] = df["issue_date"].dt.month
```

```python
df.describe()[["loan_amount", "annual_income", "dti", "loan_age_months"]]
```

```python
import os
os.makedirs('data/processed', exist_ok=True)
df.to_csv('data/processed/financial_loan_clean.csv', index=False)
```
